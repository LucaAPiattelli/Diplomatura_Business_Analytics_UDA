### Listas de textos

```
lista1 = ["Me","encanta","Python","version","3.7.10"]
print(lista1)
```

Se puede acceder a un elemento de una lista indicando su índice, es decir su posición en la lista. Los índices en Python comienzan a contar desde cero

Para acceder al tercer elemento (Python) y al quinto elemento (3.7.10) de lista, entonces:

```
print(lista1[2])
print(lista1[4])
```

Python proporciona una API para acceder a una variedad de elementos utilizando el operador : 

Solo necesita especificar el índice inicial en el lado izquierdo del operador y el índice final en el lado derecho. El índice final siempre se excluye del rango. 

Entonces, si desea obtener tres elementos (índice 2 a 4), debe hacer lo siguiente:


```
print(lista1[2:5])
```

También puede recorrer cada elemento de una lista mediante un bucle. Si desea imprimir todos los elementos de la lista, debe hacer esto:



```
for palabra in lista1:
    print(palabra)
```

Para agregar un elemento al final de la lista se usa el metodo .append()



```
lista1.append("muchisimo")
```

```
for palabra in lista1:
    print(palabra)
```

Para eliminar un elemento de la lista, se usa el metodo .remove()

```
lista1.remove("version")
```

```
for palabra in lista1:
    print(palabra)
```

## Diccionarios

```
diccionario1 = {"Modulo": "Modulo 7", "Tema": "Introducción a Python"}
print(diccionario1)
```

Para acceder a un valor específico, debe proporcionar el nombre de clave correspondiente. 

Por ejemplo, si se desea obtener el valor Introduccion a Python, entonces:

```
diccionario1["Tema"]
```

Cada par clave del valor de un diccionario debe ser único.

Python proporciona un método para acceder a todos los nombres de las claves de un diccionario. 

El metodo .keys(), que se utiliza como se muestra a continuación:

```
diccionario1.keys()
```

Tambien se puede recorrer todos los elementos de un diccionario mediante un bucle y el metodo .items(), como se muestra en el siguiente código:



```
for clave, valor in diccionario1.items():
    print(clave)
    print(valor)
```

Se puede agregar un nuevo elemento en un diccionario proporcionando el nombre de clave como este:

```
diccionario1["Profesores"] = "Machin - Ortega"
print(diccionario1)
```

Se puede eliminar un elemento de un diccionario con el comando del:

```
del diccionario1["Profesores"]
print(diccionario1)
```

## La sentencia if

Vamos a interactuar solicitando agregar un valor, positivo o negativo

```
x = int(input("Ingresar un número entero positivo o negativo, por favor: "))
```

```
if x < 0:
  print("saldo negativo") 
elif x == 0:
  print("cero")
else:
  print("saldo positivo")
```

## La sentencia for

## Definiendo funciones

La palabra reservada **def** se usa para definir funciones. Debe seguirle el nombre de la función y la lista de parámetros formales entre paréntesis. Las sentencias que forman el cuerpo de la función empiezan en la línea siguiente, y deben estar con sangría.

```
def num(x):
  if x < 0:
    print("saldo negativo") 
  elif x == 0:
    print("cero")
  else:
    print("saldo positivo")
```

```
num(5)
```

```
num(0)
```

```
num(-9)
```

```
num
x = num
```

```
x(59)
```

#### Modulo pickle

El modulo pickle implementa protocolos binarios para serializar y deserializar una estructura de Python. ***Pickling*** es el proceso mediante el cual una jerarquía de objetos de Python se convierte en una secuencia de bytes, y el ***unpickling*** es la operación inversa, mediante la cual una secuencia de bytes de un archivo binario.

El formato de datos utilizado por pickle es específico de Python. Esto tiene la ventaja de que no hay restricciones impuestas por estándares externos como por ejemplo JSON o CSV

```
import pickle
```

#### Modulo SQLite3

SQLite es una biblioteca de C que provee una base de datos ligera basada en disco que no requiere un proceso de servidor separado y permite acceder a la base de datos usando una variación no estándar del lenguaje de consulta SQL.  También es posible prototipar una aplicación usando SQLite y luego transferir el código a una base de datos más grande como PostgreSQL u Oracle

```
import sqlite3
```
