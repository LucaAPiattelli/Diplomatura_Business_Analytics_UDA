## Carga de módulos necesarios

Al cargarse el módulo Numpy debe usarse un alias de importación para poder referirse al módulo, usualmente se utiliza “np”. 

```
import numpy as np
```

Incorpora una nueva clase de objetos llamados arrays que permite representar colecciones de datos de un mismo tipo en varias dimensiones, y funciones muy eficientes para su manipulación

Un array es una estructura de datos de un mismo tipo organizada en forma de tabla o cuadrícula de distintas dimensiones. Las dimensiones de un array también se conocen como ejes.

## Dimensión de array

Para una lista de valores se crea un array de una dimensión, también conocido como vector

Captura de Pantalla 2021-07-30 a la(s) 16.54.13.png

```
a1 = np.array([7, 2, 9, 10])
print(a1)
```

```
a1.shape
```

```
a1.ndim
```

Para una lista de listas de valores se crea un array de dos dimensiones, también conocido como matriz

Captura de Pantalla 2021-07-30 a la(s) 16.54.22.png

```
a2 = np.array([[5.2, 3.0, 4.5], [9.1, 0.1, 0.3]])
print(a2)
```

```
a2.shape
```

```
a2.ndim
```

Para una lista de listas de listas de valores se crea un array de tres dimensiones, también conocido como cubo

## Operaciones

### Matematicas

Los operadores mamemáticos se utilizan para la realizar suma, resta, producto, cociente y potencia a nivel de elemento del array

```
a4 = np.array([[14, 6, 8, 16],[2, 4, 5, 6]])
a5 = np.array([[7, 2, 4, 8],[1, 2, 1, 3]])
```

```
print(a4 + a5)
```

```
print(a4 - a5)
```

```
print(a4 / a5)
```

```
print(a4 * a5)
```

### Matriciales

```
print(a1.dot(a1)) # producto matricial
```

```
print(a4.T) # transposición matricial
```
