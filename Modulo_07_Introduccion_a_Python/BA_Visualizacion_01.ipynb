# Visualizaciones de Datos - Librería MatPlotLib

## Carga de librerías necesarias

```
import pandas as pd
import matplotlib.pyplot as plt
```

## Extracción de Datos - Creación del DataFrame

```
datos_url = "https://raw.githubusercontent.com/cristiandarioortegayubro/BA/main/Datasets/Ventas"
```

```
ventas = pd.read_csv(datos_url) #los datos están separados por comas
ventas.head() #vista del dataframe
```

```
ventas.info()
```

## Formatos disponibles 
Podemos visualizar cuales son los formatos disponibles y de esa forma elegir uno, para establecer el aspecto de nuestro gráfico

```
print(plt.style.available)
```

## Graficos

### Grafico Lineal

```
plt.style.use("ggplot")
```

```
ventas.plot("FECHA","IMPORTEMONTRANSACCION", 
          kind="line", 
          figsize=(12,6), 
          color="darkblue", 
          title="ANÁLISIS DE VENTAS", 
          ylabel = "PESOS",
          xlabel = "VENTAS")
```

```
ventas.plot("FECHA",["PRECIO","IMPORTEMONTRANSACCION"], 
          kind="line", 
          figsize=(12,6), 
          color=["GREEN","darkblue"], 
          title="ANÁLISIS DE VENTAS", 
          ylabel = "PESOS",
          xlabel = "VENTAS")
```

Gráfico dividido en dos en forma horizontal

```
fig, axs = plt.subplots(2, 1, figsize=(10, 6), sharex=True)

fig.suptitle("ANALISIS DE VENTAS")
fig.tight_layout()

axs[0].plot(ventas.FECHA, ventas.PRECIO)
axs[0].set_xlabel("PRECIO DE LOS PRODUCTOS")
axs[0].set_ylabel("PESOS")

axs[1].plot(ventas.FECHA, ventas.IMPORTEMONTRANSACCION)
axs[1].set_xlabel("VENTA")
axs[1].set_ylabel("PESOS")

plt.show()
```

Gráfico dividido en dos en forma vertical

```
fig, axs = plt.subplots(1, 2, figsize=(10, 3), sharey=True)

fig.suptitle("ANALISIS DE VENTAS")
fig.tight_layout()

axs[0].plot(ventas.FECHA, ventas.PRECIO)
axs[0].set_xlabel("PRECIO DE LOS PRODUCTOS")
axs[0].set_ylabel("PESOS")

axs[1].plot(ventas.FECHA, ventas.IMPORTEMONTRANSACCION)
axs[1].set_xlabel("VENTA")
axs[1].set_ylabel("PESOS")

plt.show()
```

### Grafico de Barras verticales

```
agrupacion_clientes = ventas.groupby("ORGANIZACION").sum()
```

```
agrupacion_clientes.insert(0,column="ORGANIZACION",value=agrupacion_clientes.index.values)
agrupacion_clientes.reset_index(drop=True, inplace=True)
agrupacion_clientes.head()
```

```
agrupacion_clientes.info()
```

```
agrupacion_clientes.plot("ORGANIZACION","IMPORTEMONTRANSACCION", 
          kind="bar", 
          figsize=(12,6), 
          color=["green","darkblue"], 
          title="ANÁLISIS DE VENTAS", 
          ylabel = "PESOS",
          xlabel = "VENTAS")
```

### Grafico de Barras horizontales

```
agrupacion_clientes.plot("ORGANIZACION","IMPORTEMONTRANSACCION",
          kind="barh", 
          figsize=(12,6), 
          color=["green","darkblue"], 
          title="ANÁLISIS DE VENTAS", 
          ylabel = "PESOS",
          xlabel = "VENTAS")
```

### Histogramas

```
ventas.plot("FECHA","IMPORTEMONTRANSACCION", 
          kind="hist", 
          figsize=(12,6), 
          color="green", 
          title="ANÁLISIS DE VENTAS") 
```

### Boxplots

```
ventas.plot("FECHA",["PRECIO","IMPORTEMONPRINCIPAL","IMPORTEMONTRANSACCION"], 
          kind="box", 
          figsize=(12,6), 
          color="darkblue", 
          title="ANÁLISIS DE VENTAS", 
          ylabel = "PESOS",
          xlabel = "VENTAS")
```

### Estimación de densidad del núcleo, KDE

```
ventas.plot("FECHA",["PRECIO","IMPORTEMONTRANSACCION"], 
          kind="kde", 
          figsize=(12,6), 
          color=["darkblue","purple"], 
          title="ANÁLISIS DE VENTAS", 
          ylabel = "PESOS",
          xlabel = "VENTAS")
```

### Gráfico de área

```
ventas.plot("FECHA","IMPORTEMONTRANSACCION", 
          kind="area", 
          figsize=(12,6), 
          color="yellow", 
          title="ANÁLISIS DE VENTAS", 
          ylabel = "PESOS",
          xlabel = "VENTAS")
```

### Gráfico circular

```
agrupacion_clientes.plot("","IMPORTEMONTRANSACCION",
          kind="pie", 
          figsize=(8,8), 
          labels = agrupacion_clientes.ORGANIZACION,
          legend = "",
          ylabel = "",
          title="ANÁLISIS DE VENTAS")
```

### Diagrama de Dispersión - Scatter

```
agrupacion_clientes.plot("CANTIDAD","PRECIO", 
          kind="scatter", 
          figsize=(12,6), 
          title="ANÁLISIS DE VENTAS")
```

# Visualizaciones de Datos - Librería Altair

## Carga de librerías necesarias

```
import altair as alt
```

## Graficos

### Grafico Lineal

```
alt.Chart(agrupacion_clientes).mark_line().encode(
  x="ORGANIZACION",
  y="IMPORTEMONTRANSACCION")
```

```
alt.Chart(agrupacion_clientes).mark_line().encode(
  x="ORGANIZACION",
  y="IMPORTEMONTRANSACCION",).interactive()
```

### Graficos de Barras Verticales

```
alt.Chart(agrupacion_clientes).mark_bar().encode(
  x="ORGANIZACION",
  y="IMPORTEMONTRANSACCION",).interactive()
```

```
alt.Chart(agrupacion_clientes).mark_bar().encode(
  x="ORGANIZACION",
  y="IMPORTEMONTRANSACCION",
  color="ORGANIZACION",).interactive()
  
```

```
alt.Chart(agrupacion_clientes).mark_bar(
    cornerRadiusTopLeft=9,
    cornerRadiusTopRight=9).encode(
  x="ORGANIZACION",
  y="IMPORTEMONTRANSACCION",
  color="ORGANIZACION",).interactive()
```

### Graficos de Barras Horizontales

```
alt.Chart(agrupacion_clientes).mark_bar().encode(
  x="IMPORTEMONTRANSACCION",
  y=alt.Y("ORGANIZACION",sort="x"),
  color="ORGANIZACION",).interactive()
```

### Histogramas

```
alt.Chart(agrupacion_clientes).mark_bar().encode(
  x=alt.X("IMPORTEMONTRANSACCION", bin=True, axis=None),
  y="count()",).interactive()
```

```
histograma = alt.Chart(agrupacion_clientes).mark_bar().encode(
  x=alt.X("IMPORTEMONTRANSACCION", bin=True, axis=None),
  y="count()",).interactive()

indicador = alt.Chart(agrupacion_clientes).mark_rule(color="red").encode(
    x="mean(IMPORTEMONTRANSACCION)",
    size=alt.value(5))

histograma + indicador
```

### Gráfico de área

```
alt.Chart(agrupacion_clientes).mark_area().encode(
  x="ORGANIZACION",
  y="IMPORTEMONTRANSACCION")
```

```
alt.Chart(agrupacion_clientes).mark_area().encode(
  x="IMPORTEMONTRANSACCION",
  y="ORGANIZACION")
```

### Diagrama de Dispersión - Scatter

```
alt.Chart(agrupacion_clientes).mark_circle().encode(
  alt.X("PRECIO", bin=True),
  alt.Y("IMPORTEMONTRANSACCION", bin=True),
  color="PRECIO",
  size="PRECIO"
).interactive()
```
