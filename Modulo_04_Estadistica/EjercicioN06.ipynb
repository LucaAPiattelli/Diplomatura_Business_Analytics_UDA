**Ejercicio nº6:** Una empresa de ventas de productos de terminación de obra, tiene que cubrir los bancos todos los días. En base a datos históricos (teniendo en cuenta la inflación), el promedio es de 50 (en miles de pesos), con una desviación de 7 (en miles de pesos), suponga normalidad en la distribución:

1.¿Cuál es la media poblacional, mediana poblacional y moda poblacional?¿y la desviación estándar poblacional?

La Media=Mediana=Moda=50    y  la desviación=7

2.¿Cuál es la probabilidad de que el monto a cubrir sea menor a 50?¿Y que sea mayor a 60?

```
pnorm(50,50,7)
  
1-pnorm(60,50,7)
```

3.Calcule los cuartiles poblacionales

```
qnorm(0.25,50,7)
qnorm(0.50,50,7)
qnorm(0.75,50,7)
```

4.Genere una muestra aleatoria tamaño 100 y grafique el histograma

```
muestra=rnorm(100,50,7)

h=hist(muestra,probability = T)
```

5.En base a la muestra generada estime: la media muestral, la mediana muestral, la moda muestral, la desviación estándar muestral y los cuartiles muestrales

```
mean(muestra)
median(muestra)

h$counts
h$breaks

(45+50)/2

sd(muestra)


quantile(muestra,0.25)
quantile(muestra,0.50)
quantile(muestra,0.75)
  
```
