**Ejercicio Nº24**  Se está estudiando el sueldo que gana un empleado de seguridad privada.  Realice el análisis descriptivo: Tabla de frecuencia, gráfico de barra o histograma, media muestral, mediana muestral, moda muestral, desviación muestral, cuartiles, gráfico de caja y bigote. ¿Existe algún dato atípico ? ¿La población puede tener distribución Normal?

Cargar datos

```
salarios=c(32090,20850,23250,26400,25685,20635,22650,19800,18740,16700,21700,24950,25500,26500,25200,
21250,33900,13700,23820,22440,19930,21200,23880,25490,22200,23700,26000,21830,23300,19710)
```

Histograma

```
his=hist(salarios,probability = T,
         col = "blue",
         breaks=5)
```

Armar tabla de frecuencia

```
his$breaks
```

```
clase=c("[10000-15000]","(15000,20000]","(20000-25000]","(25000,30000]","(30000,35000]")

tabla_frecuencia=data.frame(Clase=clase,frecuencia_relativa=his$counts/length(salarios),
                            frecuencia_absoluta=his$counts)
```

```
tabla_frecuencia
```

Medidas de tendencia central

Media muestral

```
mean(salarios)
```

Mediana muestral

```
median(salarios)
```

Moda muestral (simple)

```
(20000+25000)/2
```

Medidas de dispersión

```
sd(salarios)
```

Medidas de posición

```
quantile(salarios,0.25)
quantile(salarios,0.50)
quantile(salarios,0.75)
```

Gráfico de caja y bigotes

```
caja_bigote=boxplot(salarios)
```

Datos apartados o atípicos

```
caja_bigote$out
```

Teniendo en cuenta que la forma del histograma es en forma acampanada y la media muestral es similar a la mediana muestral y a la moda muestral, es posible
que la población de origen tenga distribución Normal.
