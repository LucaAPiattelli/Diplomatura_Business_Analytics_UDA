**Ejercicio Nº22 **Generar una proceso SARIMA(0,0,0)(0,1,0) usando una distribución normal, con media 0 y desviación 1

Instalar librería

```
install.packages("forecast")
```

Cargar librería

```
library(forecast)
```

1.Cada 12 meses sumar 5, a la serie

```


tiempo=c(1:(12*10))

x=rnorm(1,0,1)


t=1

while(t<=120){
  
  
  x[t]=rnorm(1,0,1)
  
  
  t=t+1  
}
```

```

plot(x, ylab="Xt", xlab="Tiempo" ,type="o",lwd=1,pch=1,cex=1,col="Green")
abline(h=0)
```

Sumar estacionalidad

```
i=1

while(i<=10){
  
  
  x[12*i]=5
  
  
  i=i+1  
}



```

2.Gráficar

```
plot(x, ylab="Xt", xlab="Tiempo" ,type="o",lwd=1,pch=1,cex=1,col="Green")
abline(h=0)
```

3.Estimar el modelo

```
#------------------------------------------------------------------------
#Crear datos ts
x = ts(x,frequency = 12)
```

```

#------------------------------------------------------------------------
fit=auto.arima(x,seasonal = T,ic = c("aicc"), stepwise=T, allowdrift = F)
summary(fit)
```
