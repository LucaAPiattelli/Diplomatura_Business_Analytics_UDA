1.Usar parametro 0,5

```
tiempo=c(1:100)

x=rnorm(1,0,1)


t=2

while(t<=100){
  

  x[t]=0.5*x[t-1]+rnorm(1,0,1)
  
  
  t=t+1  
}
```

2.Grficar

```
#-----------------------------------------------------------------------


plot(x, ylab="Xt", xlab="Tiempo" ,type="o",lwd=1,pch=1,cex=1,col="Green")
abline(h=0)
```
