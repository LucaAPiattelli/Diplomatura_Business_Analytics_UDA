**Ejercicio Nº19** Generar un ARMA(1,1) usando una distribución normal, con media 0 y desviación 1

```
#-------------------------------------------------------------------------------
#Instalar librerias
#-------------------------------------------------------------------------------

install.packages("forecast")
```

```
  #-----------------------------------------------------------------------------------------
  #Cargar librerias
  #----------------------------------------------------------------------------------------
  
  library(forecast)
```

1.Usar 0,6 para ambos parámeteros

```
#--------------------------------------------------------------------

tiempo=c(1:100)

x=rnorm(1,0,1)
e=rnorm(1,0,1)

t=2

while(t<=100){
  
  e[t]=rnorm(1,0,1) 
  x[t]=0.6*x[t-1]+0.6*e[t-1]+e[t]
  
  
  t=t+1  
}
```

2.Gráficar

```
 #-----------------------------------------------------------------------  
  
  plot(x, ylab="Xt", xlab="Tiempo" ,type="o",lwd=1,pch=1,cex=1,col="Green")
  abline(h=0)
```

3.Estimar el modelo

```
  #------------------------------------------------------------------- 
  
  
  fit=auto.arima(x,seasonal = F,ic = c("aicc"), stepwise=T, allowdrift = F)
  summary(fit)
```
