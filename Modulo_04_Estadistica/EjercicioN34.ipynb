**Ejercicio Nº34** Los clientes llaman a un call center para hacer reservaciones en un Hotel.  Se usa análisis de negocios para determinar las siguientes variables: tiempo promedio en espera antes de que el cliente tenga contacto con un representante y satisfacción del cliente con la experiencia global del call center. La empresa ha recopilado la siguiente información. ¿Qué puede decir que la satisfacción del cliente  es explicada por el tiempo de espera? La satisfacción del cliente está en porcentaje, el decir 96 significa que el 96% quedó satisfecho. Usar el modelo lineal simple


Cargar librería

```r
library(ggplot2)
```

Cargar datos

```r
satisfaccion_cliente=c(96,92,82,84,90,80,74)
tiempo_espera=c(22,34,44,67,38,70,86)
```

Gráficar

```r
g=ggplot(data.frame(satisfaccion_cliente=satisfaccion_cliente,tiempo_espera=tiempo_espera),aes(x=tiempo_espera,y=satisfaccion_cliente))
g=g+geom_smooth(method=lm,formula = y ~ poly(x, 1))
g=g+geom_point(shape=1)
g
```

Estimamos modelo lineal

```r
modelo=lm(satisfaccion_cliente~tiempo_espera)
summary(modelo)
```

Tanto la puebra t como la F nos indican que hay evidencia para rechazar la no relación entre la variables. Por otro lado el coeficiente de determinación indica que la satisfaccion del cliente es explicado en un 86% por el tiempo de espera. Resumiento hay evidencia para considerar la satisfacción del cliente es explicado en parte por el tiempo de espera.
