**Ejercicio Nº17**: Generar un MA(1) usando una distribución normal, con media 0 y desviación de 1

Instalar librería

```
install.packages("forecast")
```

Cargar librería

```

library(forecast)
```

1.Usar parámetro 0,5

```
#--------------------------------------------------------------------

tiempo=c(1:100)

x=rnorm(1,0,1)
e=rnorm(1,0,1)

t=2

while(t<=100){
  
  e[t]=rnorm(1,0,1) 
  x[t]=0.5*e[t-1]+e[t]
  
  
  t=t+1  
}
```

2.Gráficar

```
#-----------------------------------------------------------------------


plot(x, ylab="Xt", xlab="Tiempo" ,type="o",lwd=1,pch=1,cex=1,col="Green")
abline(h=0)
```

3.Estimar el modelo


```
fit=auto.arima(x,seasonal = F,ic = c("aicc"), stepwise=T, allowdrift = F)
summary(fit)
```
