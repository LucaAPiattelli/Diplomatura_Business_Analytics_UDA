**Ejercicio Nº32** Las existencias agotadas ocurren cuando la demanda del tiempo de entrega sobrepasa el punto de reposición. Cuando la demanda es incierta, usar el modelo EOQ a partir sólo de la demanda promedio resultará en una alta probabilidad de existencia agotada, ya que unos de los supuestos es que la demanda es constante.
Una manera de reducir este riesgo es incrementar el punto de reposición al añadir existencias adicionales. Las existencias de seguridad son inventarios planeados y adicionales a la mano que actúan como reservas para reducir el riesgo de existencia agotada.
Para determinar el punto de reposición r, se necesita determinar la distribución de probabilidad de la demanda, que generalmente se supone normal. El punto de reposición depende del riesgo que se quiera asumir por existencias agotadas.

El dueño de la librería del ejercicio anterior, quiere determinar el punto de reposición, para cumplir por lo menos el 95% de las veces, sabiendo que cuando hace el pedido la pastera tarda una semana en hacer que las lleguen a la librería.  Para ello registra la cantidad demanda de resmas venidas por semana:

1.En base al análisis descriptivo determine si se puede considerar que esta muestra viene de una población normal

Podemos hacer el histograma y gráficar la campana  usando la media muestral estimada y la desviación muestral estimada. De esta forma vemos si hay coicidencia entre las barras y la curva!!!

```r
hist=hist(demanda_semanal,probability = T,col = "green")


curve(dnorm(x, mean=mean(demanda_semanal), sd=sd(demanda_semanal)), 
       col="darkblue", lwd=2,
       add=TRUE, yaxt="n")
```

Calculamos el valor de z, que es un cuantil. Ya que se quiere cumplir con el 95%, entonces el centro debe ser 0,95 y 0,025 en cada uno de los extremos. Es decir para 0,025 ó 0,975
