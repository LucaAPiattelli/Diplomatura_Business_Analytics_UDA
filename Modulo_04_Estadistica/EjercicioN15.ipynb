**Ejercicio Nº 15** La administración de una empresa de ventas de purificadores de aire, está planeando, pagar un porcentaje del sueldo en función de la antigüedad del empleado.  Esto se justifica si el empleado que tiene más antigüedad en la empresa, debería  tener  más experiencia y por lo tanto mayores ventas. Para ello se tomó una muestra de 40 empleados, en donde se midió antigüedad en meses y la cantidad de purificadores vendidos.

Instalar Librería

```
install.packages("lmtest")
```

Cargar librería

```
library(ggplot2)
library(lmtest)
```

1.Graficar antiguedad vs ventas, forma lineal y polinómica de grado 2

Cargar datos

```
antiguedad=c(122,170,151,140,112,62,79,100,177,5,158,173,152,127,49,2,58,52,103,35,155,133,59,195,50,68,137,
            196, 167, 149, 108, 142,  23,  25, 116,  11,  94, 165, 107, 179)



ventas=c(95,82,98,101,90,54,64,94,95,1,87,81,115,98,62,28,53,37,87,30 ,88 ,74,82,100,72 ,67 ,94,
          93,83,95,98,74,54,27,74,27,81,86,85,94)
  
   
```

Gráficar

```


g=ggplot(data.frame(antiguedad=antiguedad,ventas=ventas),aes(x=antiguedad,y=ventas))
g=g+geom_smooth(method=lm,formula = y ~ poly(x, 2))
g=g+geom_point(shape=1)
g 
```

2.Estimar el modelo de regresión lineal y polinómica de grado 2 ¿Cuál tiene el mayor coeficiente de determinación?

```
modelo=lm(ventas~poly(antiguedad,2))

summary(modelo)
    
```

3.Analizar los supuestos del modelo

Test de shapiro-Wilks

H0: Es normal    vs    H1: no es normal

```
shapiro.test(rstandard(modelo))

hist(modelo$residuals)

```

Test Breusch-Pagan test

H0: todas las varianzas son iguales vs H1: al menos una es diferente

```
bptest(modelo)
```

```
#--------------------------------------------------------------------------------------
#Durbin-Watson
#--------------------------------------------------------------------------------------
#H0: errores son independientes  vs H1: no son independientes

dwtest(modelo)
```
