**Ejercicio Nº29** Se sabe que la vida, en años, de las baterías que ud consume para sus camiones tiene una distribución exponencial con una vida promedio de dos años. Dada esta información calcule:

1.La probabilidad de que una batería dure menos de tres años.

```
pexp(3,1/2)
```

2.La probabilidad de que una batería dure entre 1 y 5 años.

```
pexp(5,1/2)-pexp(1,1/2)
```

3.La probabilidad de que una batería dure mas de 4 años.

```
1-pexp(4,1/2)
```

4.La probabilidad de que una batería dure mas de 4 años.

Los datos apartados son valores que estan por fuera del bigote

```
q1=qexp(0.25,1/2)
q2=qexp(0.75,1/2)

RIC=q2-q1

valor_bigote_min=q2-RIC*1.5
valor_bigote_min
valor_bigote_max=q1+RIC*1.5
valor_bigote_max
```

```
pexp(valor_bigote_min,1/2)+(1-pexp(valor_bigote_max,1/2))
```
