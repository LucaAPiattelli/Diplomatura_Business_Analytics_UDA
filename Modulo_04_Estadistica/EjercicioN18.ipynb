**Ejercicio Nº18**  Generar un AR(1) usando una distribución normal, con media 0 y desviación 1

Instalar librería

```
install.packages("forecast")
```

Cargar librería

```
library(forecast)
```

1.Usar parametro 0,5

```
tiempo=c(1:100)

x=rnorm(1,0,1)


t=2

while(t<=100){
  

  x[t]=0.5*x[t-1]+rnorm(1,0,1)
  
  
  t=t+1  
}
```

2.Gráficar

```
plot(x, ylab="Xt", xlab="Tiempo" ,type="o",lwd=1,pch=1,cex=1,col="Green")
abline(h=0)
```

3.Estimar el modelo

```
fit=auto.arima(x,seasonal = F,ic = c("aicc"), stepwise=T, allowdrift = F)
summary(fit)
```
