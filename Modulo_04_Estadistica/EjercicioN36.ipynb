**Ejercicio Nº36** En base a los datos del ejercicio Nº12, modele la relación entre cantidad de afiliados vs PBI per cápita, usando la regresión polinómica de grado 2.  Compruebe si el coeficiente de determinación es mejorado. Recuerde siempre  usar las pruebas t y F y hacer el análisis de los supuestos.

Instalar Librería

```r
install.packages("lmtest")
```

Cargar librería

```r
library(ggplot2)
library(lmtest)
```

Cargar datos

```r
pbi_pc=c(5110,5919,7245,9021,8225,10386,12849,13083,13080,12335,13789,12790,14613,11633,9912,8938)
          

afiliados=c(120,50,70,130,60,60,150,150,90,170,220,160,150,60,50,60)
```

Modelo con polinómio de grado 2

Gráficar

```r
g=ggplot(data.frame(pbi_pc=pbi_pc,afiliados=afiliados),aes(x=pbi_pc,y=afiliados))
g=g+geom_smooth(method=lm,formula = y ~ poly(x, 2))
g=g+geom_point(shape=1)
g
```

Estimar modelo

```r
modelo_grado_2=lm(afiliados~poly(pbi_pc, 2))

summary(modelo_grado_2)
```

La prueba t para la variable de grado  2, indica que hay evidencia para considerar que explica la cantidad de afiliados. Por otro lado el coeficiente de determinación se ve mejorado a un 53%.

Test de shapiro-Wilks

H0: Es normal vs H1: no es normal


```r
shapiro.test(rstandard(modelo_grado_2))
```

Test Breusch-Pagan test

H0: todas las varianzas son iguales vs H1: al menos una es diferente


```r
bptest(modelo_grado_2)
```

Durbin-Watson

H0: errores son independientes vs H1: no son independientes


```r
dwtest(modelo_grado_2)
```
