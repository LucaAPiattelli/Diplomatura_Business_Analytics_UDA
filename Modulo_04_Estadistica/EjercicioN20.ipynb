**Ejercicio Nº20** Generar una proceso con tendencia determinística usando una distribución normal, con media 0 y desviación 1

Instalar librería

```
install.packages("forecast")
```

Cargar librería

```
  library(forecast)
```

1.Usar  y  ordenada al origen 0,1 y pendiente 0,1

```
tiempo=c(1:100)

x=rnorm(1,0,1)


t=1

while(t<=100){
  
  
  x[t]=0.1+0.1*tiempo[t]+rnorm(1,0,1)
  
  
  t=t+1  
}
```

2.Gráficar

```
plot(x, ylab="Xt", xlab="Tiempo" ,type="o",lwd=1,pch=1,cex=1,col="Green")
abline(h=0)
```

3.Estimar el modelo

```
fit=auto.arima(x,seasonal = F,ic = c("aicc"), stepwise=T, allowdrift = T)
summary(fit)
```

Se puede estimar el modelo diferenciando directamente

```
dx=diff(x,1)


plot(dx, ylab="dXt", xlab="Tiempo" ,type="o",lwd=1,pch=1,cex=1,col="Green")
abline(h=0)
```

```
fit=auto.arima(dx,seasonal = F,ic = c("aicc"), stepwise=T, allowdrift = F)
summary(fit)




```
