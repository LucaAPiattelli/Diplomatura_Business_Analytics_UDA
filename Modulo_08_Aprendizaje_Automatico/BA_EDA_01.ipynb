# **Carga de módulos y datos necesarios**

```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

Antes que nada, se genera el dataframe con los datos que se utilizaran para la visualización


```
url = "https://raw.githubusercontent.com/LucaAPiattelli/Diplomatura_Business_Analytics_UDA/main/Modulo_08_Aprendizaje_Automatico/visualizacion.csv"
```

```
analisis = pd.read_csv(url)
analisis.head()
```

Se cambian los nombres de las columnas para mejor interpretacion

```
analisis.info()
```

```
analisis.rename(columns={"satisfaction_level":"niveldesatisfaccion",
                         "last_evaluation":"ultimaevaluacion",
                         "number_project":"numerosdeproyectos",
                         "average_montly_hours":"horasmensualespromedio",
                         "time_spend_company":"tiempoenlaempresa",
                         "Work_accident":"accidentedetrabajo",
                         "left":"abandono",
                         "promotion_last_5years":"promocionultimos5años",
                         "sales":"ventas",
                         "salary":"sueldo"}, inplace= True)
```

```
analisis.info()
```

# **Visualización de los datos**

```
fig1 = px.scatter(analisis, x="niveldesatisfaccion", 
                y="ultimaevaluacion",
                color="abandono")
fig1.show()
```

```
fig2 = px.scatter(analisis, x="niveldesatisfaccion", 
                y="ultimaevaluacion",
                color=analisis.abandono.apply(str))
fig2.show()
```

```
fig1 = px.scatter(analisis, x="niveldesatisfaccion", 
                y="ultimaevaluacion",
                color_continuous_scale="mint",
                template="gridon",
                color="abandono",                   
                labels={"niveldesatisfaccion":'Nivel de Satisfacción',
                        "ultimaevaluacion":"Ultima Evaluación",
                        "abandono":"Abandono"})

fig1.show()
```

```
fig = px.histogram(analisis, 
                   barmode="relative",
                   title="Nivel de Satisfacción",
                   template="gridon",
                   x="niveldesatisfaccion",
                   labels={"niveldesatisfaccion":"Nivel de Satisfacción"})

fig.update_layout(bargap=0.2)
fig.show()
```

```
fig = px.histogram(analisis, 
                   barmode="relative",
                   title="Ultima Evaluación",
                   template="gridon",
                   x="ultimaevaluacion",
                   labels={"ultimaevaluacion":"Ultima Evaluación"})

fig.update_layout(bargap=0.2)
fig.show()
```

```
fig = px.histogram(analisis, x="sueldo",
                   title="Sueldos",
                   color="sueldo",
                   template="gridon")

fig.show()
```

```
fig = px.histogram(analisis, x="sueldo",
                   title="Sueldos",
                   color="abandono",
                   template="gridon")

fig.show()
```

```
fig1 = px.scatter(analisis, x="niveldesatisfaccion", 
                y="ultimaevaluacion",
                color="abandono",
                template="gridon",
                marginal_x="histogram",
                marginal_y="histogram",
                labels={"niveldesatisfaccion":'Nivel de Satisfacción',
                        "ultimaevaluacion":"Ultima Evaluación"})

fig1.show()
```

```
fig1 = px.scatter(analisis, x="niveldesatisfaccion", 
                y="ultimaevaluacion",
                color="abandono",
                template="gridon",
                marginal_x="violin",
                marginal_y="violin",
                labels={"niveldesatisfaccion":'Nivel de Satisfacción',
                        "ultimaevaluacion":"Ultima Evaluación"})

fig1.show()
```
